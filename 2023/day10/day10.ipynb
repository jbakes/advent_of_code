```python
# Pad row
l = ''
maze = []
for i in range(len(data[0])):
    l = l + '.'
maze.append('.' + l+ '.')
# data.insert(0,l)
# data.append(l)
for i in range(len(data)):
    maze.append('.' + data[i] + '.')
maze.append('.' + l + '.')
```

```python
for i in range(len(maze)):
    for j in range(len(maze)):
        if maze[i][j] == 'S':
            start = (i,j)
start
```

```python
def find_path(i_loc, j_loc, history):
    if maze[i_loc - 1][j_loc] in ('|', '7', 'F') and maze[i_loc][j_loc] in ('S', '|', 'L', 'J'):
        if (i_loc -1, j_loc) not in history:
            return (i_loc -1, j_loc)
    if maze[i_loc + 1][j_loc] in ('|', 'L', 'J') and maze[i_loc][j_loc] in ('S', '|', '7', 'F'):
        if (i_loc + 1, j_loc) not in history:
            return (i_loc + 1, j_loc)
    if maze[i_loc][j_loc - 1] in ('-', 'L', 'F') and maze[i_loc][j_loc] in ('S', '-', '7', 'J'):
        if (i_loc, j_loc - 1) not in history:
            return (i_loc, j_loc - 1)
    if maze[i_loc][j_loc + 1] in ('-', '7', 'J') and maze[i_loc][j_loc] in ('S', '-', 'L', 'F'):
        if (i_loc, j_loc + 1) not in history:
            return (i_loc, j_loc + 1)
    return (-1, -1)
        
```

```python
next_spot = start
history = [start]

while True:
    next_spot = find_path(next_spot[0], next_spot[1], history)
    if next_spot == (-1, -1):
        break
    history.append(next_spot)
```
